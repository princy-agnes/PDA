#FUNCTIONS

```python
def fun():
    print("welcome")
fun()
```

#DEFAULT ARGUMENT

```python
def myfun(x,y=20):
    print("x:",x)
    print("y:",y)
myfun(30)
```

#ASSIGNMENT PASSING

```python
def evenodd(x):
    if x%2==0:
        print("Even")
    else:
        print("Odd")
print(evenodd(16))
print(evenodd(15))
```

#KEYWORD ARGUMENT

```python
def student(fname,lname):
    print(fname,lname)
student(fname = "practice",lname = "daily")
student(lname = "daily",fname = "pratice")
```

#POSITIONAL ARGUMENT

```python
def person(name,age):
    print("hi I am",name)
    print("hi I am",age)
print("case 1")
person("Ram",25)
print("case 2")
person("25","Ram")
```

#FUNCTION WITHIN FUNCTION

```python
def f1():
    s = "I love India"
    def f2():
        print(s)
    f2()
f1()
```

#PASS BY REFERENCE

```python
def myfun(x):
    x[0]=20
lst = [10,20,30,40,50]
print(lst)
myfun(lst)
print(lst)
```

#PASS BY VALUE

```python
def myfun(x):
    x=20
a=20
print(a)
myfun(a)
print(a)
```
